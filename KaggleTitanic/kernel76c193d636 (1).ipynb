```python
y_train=trainset['Survived'].copy()
trainset=trainset.drop(['Survived'],1)

#combining train and test dataset...
combine=pd.concat([trainset,testset])


#droping unuseful columns..
combine=combine.drop(['Name','PassengerId','Ticket'],1)

combine
```

```python
#Label encoding Sex
import sklearn.preprocessing as pp
lb=pp.LabelBinarizer()
combine.Sex=lb.fit_transform(combine.Sex)


#checking for null entries
combine.isnull().sum()
```

```python
#filling null values for AGE , by median value

combine['Age'].fillna(combine['Age'].median(), inplace=True)
combine.isnull().sum()


 
#filling the null value for FARE , by mean 
# combine['Fare'].fillna(combine['Fare'].mean(), inplace=True)

# fill missing Fare with median fare for each Pclass
combine['Fare'].fillna(combine.groupby("Pclass")["Fare"].transform("median"), inplace=True)



# filling CABIN null entries by 'U' as unknown 
    
    
   #combine['Cabin'].value_counts()   #lists the all entries along with their number of occurence

combine['Cabin']=combine['Cabin'].fillna('U')              # U for Unknow category
combine['Cabin']=combine['Cabin'].astype(str).str[0]       #Modifying entries ,just placing the initial letters of the string 



#filling most frequent value in the EMBARKED  null entries 
combine['Embarked']=combine['Embarked'].fillna( combine['Embarked'].value_counts().index[0])

combine

```

```python
#mapping for a range 

combine["FamilySize"] = combine["SibSp"] + combine["Parch"] + 1

combine['Singleton'] = combine['FamilySize'].map(lambda s: 1 if s == 1 else 0)
combine['SmallFamily'] = combine['FamilySize'].map(lambda s: 1 if 2<=s<=4 else 0)
combine['LargeFamily'] = combine['FamilySize'].map(lambda s: 1 if 5<=s else 0)
combine.head(10)
```

```python
combine=combine.drop(['FamilySize'],1)

#label Encodeing 

from sklearn.preprocessing import LabelEncoder
le_embarked = LabelEncoder()
combine['Embarked'] = le_embarked.fit_transform(combine['Embarked'])


le_cabin = LabelEncoder()
combine['Cabin'] = le_cabin.fit_transform(combine['Cabin'])

combine=combine.drop(['SibSp', 'Parch'] ,1)
combine.head(20)


X_train = combine[0:len(trainset)]
X_test = combine[len(trainset):]
```

```python

# Importing Classifier Modules
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

 
target = y_train
features = X_train.values

clf = DecisionTreeClassifier()
clf_ = clf.fit(features, target) 

print(clf_.score(features, target))
```

```python
from sklearn import metrics
y_pred = clf.predict(X_train)

print("Accuracy:",metrics.accuracy_score(y_train, y_pred))
```
